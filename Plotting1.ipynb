`f(x) = `_`some expression`_

Function names can generally be any combination of letters and numbers (but not starting with a number). So we could create a function called `f(x)` or `f1(x)` or `Temp(x)`, etc.

Sage uses the typical operation symbols `+`, `-`, `*`, `/`, `^`; names like `pi` and `e`; and many common functions `sin`, `log`, etc.

To plot a function, you would use SageMath's `plot()` method. This can be used in two different ways

`plot(f)`

Suppose we wanted to narrow in on where the function crosses the x-axis (in the example $f(x) = e^x - 3x$ it seems to be between 0.6 and 0.7). The plot() method provides two ways to specify a range of values for the x-axis to use.

One way is to include an order triple with x, the lower limit, and the upper limit with its parts

`plot(f, (x, 0.5, 0.6))`

## Part 2 - Graphing Multiple Functions

The plot() method can be used to graph two functions on the same axis. Suppose we define a second function

`g(x) = x * cos(x) + 2` (for $g(x) = x\cos x + 2$)

5. Create a second function g



To graph both functions, include them as an ordered pair in the parentheses for plot()

`plot((f, g))`

6. Plot both of your functions

The default color for the first plot is blue, and the second is green. But you can changes those if you wish.

You can even specify a range for x to zoom in on where the graphs

`plot((f, g), (x, -0.4, -0.2))`
